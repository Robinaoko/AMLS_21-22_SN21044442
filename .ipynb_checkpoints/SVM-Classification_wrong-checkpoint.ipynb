```python
import numpy as np
import os
from pathlib import Path

import matplotlib.pyplot as plt
from keras.preprocessing import image
```

```python
p = Path("images/")
dirs = p.glob("*")

image_data = []
   

for folder_dir in dirs:
    
    for img_path in folder_dir.glob("*.jpg"):
        img = image.load_img(img_path, target_size=(100,100))
        img_array = image.img_to_array(img)
        image_data.append(img_array)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tumor_label=pd.read_csv('label.csv')
label = []
for i in range(3000):
    data = tumor_label['label'][i]
    if data == 'no_tumor':
         label.append(data)
    else :
         label.append('yes_tumor')
```

```python
label[0 : 10 ]
```

```python
#X_train, X_test, y_train, y_test = train_test_split(image_data,label, test_size=0.3,random_state=109)
```

```python
image_data = np.array(image_data, dtype='float32')/255.0
label = np.array(label)

print(image_data.shape, label.shape)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.imshow(image_data[3])
plt.axis('off')
plt.show()
```

```python
M = image_data.shape[0]
image_data = image_data.reshape(M,-1)
print(image_data.shape)
print(label.shape)
```

```python
from sklearn import svm

svm_classifier = svm.SVC(kernel='linear', C=1.0)
svm_classifier.fit(image_data, label)
ypred_sklearn = svm_classifier.predict(image_data)
svm_classifier.score(image_data,label)
```
